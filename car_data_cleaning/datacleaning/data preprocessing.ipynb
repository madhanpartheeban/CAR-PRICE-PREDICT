```python
import re

def convert_price(price_str):
    # Remove ₹ symbol and commas (if any)
    price_str = re.sub(r'[₹,]', '', price_str)

    # Strip any leading or trailing whitespace
    price_str = price_str.strip()

    # Handle 'Lakh' or 'Lakhs'
    if 'Lakh' in price_str:
        # Remove 'Lakh' or 'Lakhs' and convert the remaining string to a float
        price_str = price_str.replace('Lakh', '').replace('Lakhs', '').strip()
        return float(price_str) * 10**5

    # Handle 'Crore' or 'Crores'
    elif 'Crore' in price_str:
        # Remove 'Crore' or 'Crores' and convert the remaining string to a float
        price_str = price_str.replace('Crore', '').replace('Crores', '').strip()
        return float(price_str) * 10**7

    # For values without 'Lakh' or 'Crore', just return the float version of the number
    else:
        return float(price_str)

# Example usage
df['price'] = df['price'].apply(convert_price)
```

```python
# changing the dtype and removing the seats string in columns
df['Seats'] = df['Seats'].str.replace('Seats', '').astype(float)
```

```python
# changing the d type and removing the string in enhine columns

df['Engine'] = df['Engine'].str.replace('CC','').astype(float)
```

```python
#changing dtype and removing string value in mileage column and topspeed
df['Mileage'] = df['Mileage'].str.replace(' kmpl', '').str.replace('km/kg','').astype(float)
```

```python
df['Top Speed'] = df['Top Speed'].str.replace(r'[^\d.]', '', regex=True).replace('',np.nan)
df['Top Speed'] = df['Top Speed'].astype(float)
```
